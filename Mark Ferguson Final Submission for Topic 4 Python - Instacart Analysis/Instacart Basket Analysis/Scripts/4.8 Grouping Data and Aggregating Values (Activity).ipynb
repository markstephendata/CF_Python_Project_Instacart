# Create Loyalty Flag

```python
# Question 4
```

```python
#execute command to create column 'max_order' to establish number of orders per user

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
#Define Loyalty Criteria
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Test the distribution of newly created category
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# have a look to see if the max_order values match up with the loyalty_flag criteria...

ords_prods_merge[['user_id', 'loyalty_flag', 'max_order']].head(60)
```

```python
# Question 5

# Using the new categories of loyalty_flag, I aggregated by the 'prices' column 
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Question 6

# create low and high spender categories where the average price per item is <10 or >= 10 respectively
```

```python
# Creating “average_price” column using transform()

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Creating a 'spending_flag' flag (Low spender)

ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Creating a 'spending_flag' flag (High spender)

ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check the column and the flag are displayed correctly:

ords_prods_merge.tail(50)
```

```python
# Check the frequency

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Question 7

# Classify 'frequent' and 'non-frequent' customers
# Use each customer's median of 'days_since_prior_order' column to define
# >20 'Non frequent customer'
# >10, <=20 'Regular customer'
# <=10 'Frequent customer'
```

```python
ords_prods_merge.head(15)
```

```python
# Creating “median_prior_orders” column using transform()

ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating a 'order_frequency_flag' flag

ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating a 'order_frequency_flag' flag

ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] > 10) & (ords_prods_merge['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Creating a 'order_frequency_flag' flag

ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
# Check the frequency

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
